## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price)
plt.plot(days, predicted_price)
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

```python
plt.plot(days, original_price, 'b', label='Original Price')
plt.plot(days, predicted_price, 'r', label='Predicted Price')
plt.legend()
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
```

-------------------------------------------------------------------------------------------------------------------------------------------------------

### Subplots

```python
# First subplot

plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

# Second subplot
plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted prices over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();


# Third subplot - Difference between prices
price_difference = [o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days, price_difference, color='purple', label='Price difference')
plt.title('Difference between original and predicted price');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend();

plt.tight_layout()
```
